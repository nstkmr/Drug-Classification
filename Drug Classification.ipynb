```python
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\Nishant\\Machine Learning\\Datasets\\Drug Classification.csv")
df.head()
```

```python
dummies = pd.get_dummies(df.Sex).astype(int)
dummies.head()
```

```python
dummies2 = pd.get_dummies(df.BP).astype(int)
dummies2.head()
```

```python
dummies3 = pd.get_dummies(df.Cholesterol).astype(int)
dummies3.head()
```

```python
dummies3.rename(columns={'HIGH': 'HIGH1', 'NORMAL': 'NORMAL1'}, inplace=True)
dummies3.head()
```

```python
df['Drug'] = df['Drug'].replace({'drugA': 0, 'drugB': 1, 'drugC': 2, 'drugX': 3, 'drugY': 4}).astype(int)
df.head()
```

```python
df1 = pd.concat([df,dummies,dummies2,dummies3],axis='columns')
df1.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
cont_cols = ['Age', 'Na', 'K']
df1[cont_cols] = scaler.fit_transform(df1[cont_cols])
df1.head()
```

```python
X = df1.drop(['Sex','BP','Cholesterol','Drug','M','NORMAL','NORMAL1'],axis='columns')
X.head()
```

```python
y = df1['Drug']
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(len(X_train))
print(len(X_test))
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model_C = SVC(C=100)
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)
```

```python
model_g = SVC(gamma=0.3)
model_g.fit(X_train, y_train)
model_g.score(X_test, y_test)
```

```python
model_linear_kernal = SVC(kernel='linear')
model_linear_kernal.fit(X_train, y_train)
```

```python
model_linear_kernal.score(X_test, y_test)
```
